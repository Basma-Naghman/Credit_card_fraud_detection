# Under Sampling

## Build the sample dataset containing simular distribution of normal transaction and fraudulent transaction

### Fraudulent transaction ---> 492

```python
legit_sample = legit.sample(n=492)
```

## Concatenate two dataframe

```python
new_df = pd.concat([legit_sample , fraud] , axis= 0)
```

```python
new_df.head()
```

```python
new_df["Class"].value_counts()
```

```python
new_df.groupby("Class").mean()
```

# Splitting

```python
x = new_df.drop(columns=["Class"] , axis = 0)
```

```python
y = new_df["Class"]
```

```python
x
```

```python
y
```

```python
x_train , x_test  , y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
print("Shape of dataset" , new_df.shape)
print("Shape of x_train", x_train.shape)
print("Shape of x_test", x_test.shape)
print("Shape of y_train", y_train.shape)
print("Shape of y_test", y_test.shape)
```

# Model training

```python
# Logistic Regression
model = LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

# Model Evaluation

```python
 y_predict = model.predict(x_test)
 y_predict
```

# Accuracy

```python
accuracy_score = accuracy_score(y_test, y_predict)
```

```python
print("Accuracy on test data " , accuracy_score)
```
